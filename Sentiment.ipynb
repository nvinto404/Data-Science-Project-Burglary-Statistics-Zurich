```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.zh.ch/de/news-uebersicht.html?organisation=organisationen%253Akanton-zuerich%252Fsicherheitsdirektion%252Fkantonspolizei&fullText=Einbruch&page=1&orderBy=new'

try:
    response = requests.get(url)
    response.raise_for_status()  # Überprüfe, ob der Seitenabruf erfolgreich war

    soup = BeautifulSoup(response.text, 'html.parser')

    for link in soup.select('.mdl-news-teaser__item a'):
        article_title = link.select_one('h1.atm-heading').get_text(strip=True)
        article_date = link.select_one('p.mdl-lead_section__dateline').get_text(strip=True)
        article_text = link.select_one('.cell > div > div > div.text p').get_text(strip=True)

        print("Titel:", article_title)
        print("Datum:", article_date)
        print("Text:", article_text)
        print("\n")

except requests.exceptions.RequestException as e:
    print("Fehler beim Seitenabruf:", e)
```

```python
import requests
from bs4 import BeautifulSoup

# URL der Hauptseite
main_url = 'https://www.zh.ch/de/news-uebersicht.html?organisation=organisationen%253Akanton-zuerich%252Fsicherheitsdirektion%252Fkantonspolizei&fullText=Einbruch&page=1&orderBy=relevance'

# Eine GET-Anfrage an die Hauptseite senden und den HTML-Inhalt abrufen
main_response = requests.get(main_url)
main_soup = BeautifulSoup(main_response.text, 'html.parser')

# Selektor für alle Tiles auf der Hauptseite
tiles_selector = 'ul.mdl-news-list__items li a'

# Alle Tiles auf der Hauptseite auswählen
tiles = main_soup.select(tiles_selector)

# Durch alle Tiles iterieren
for tile in tiles:
    # URL des aktuellen Tiles extrahieren
    tile_url = tile['href']
    
    # Eine GET-Anfrage an die URL des aktuellen Tiles senden und den HTML-Inhalt abrufen
    tile_response = requests.get(tile_url)
    tile_soup = BeautifulSoup(tile_response.text, 'html.parser')
    
    # Titel des Artikels extrahieren
    title_element = tile_soup.select_one('h1.atm-heading')
    article_title = title_element.get_text(strip=True) if title_element else "Titel nicht gefunden"
    
    # Datum des Artikels extrahieren
    date_element = tile_soup.select_one('div.mdl-lead_section__date time')
    article_date = date_element.get_text(strip=True) if date_element else "Datum nicht gefunden"
    
    # Inhalt des Artikels extrahieren
    content_element = tile_soup.select_one('div.mdl-news-article__content p')
    article_content = content_element.get_text(strip=True) if content_element else "Inhalt nicht gefunden"
    
    # Die extrahierten Informationen ausgeben
    print("Titel:", article_title)
    print("Datum:", article_date)
    print("Inhalt:", article_content)
    print("\n")
```

```python
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException

# Initiiere den Chrome Webdriver mit dem ChromeDriverManager
driver = webdriver.Firefox()

# URL der Hauptseite
main_url = 'https://www.zh.ch/de/news-uebersicht.html?organisation=organisationen%253Akanton-zuerich%252Fsicherheitsdirektion%252Fkantonspolizei&fullText=Einbruch&page=1&orderBy=relevance'

# Lade die Hauptseite
driver.get(main_url)

# Warte, bis die Seite vollständig geladen ist (hier warten wir maximal 10 Sekunden)
driver.implicitly_wait(10)

# Selektor für alle Tiles auf der Hauptseite
tiles_selector = '/html/body/main/div/div/div[2]/div/div/div/div/div/div[2]/div[1]/ul/li/a'

# Durch alle Tiles iterieren
while True:
    try:
        # Alle Tiles auf der Hauptseite auswählen
        tiles = driver.find_elements("xpath", tiles_selector)
        
        # Wenn keine Tiles mehr gefunden werden, breche die Schleife ab
        if not tiles:
            break

        # Iteriere über die gefundenen Tiles
        for tile in tiles:
            try:
                # Klicke auf das aktuelle Tile, um zur Artikelseite zu gelangen
                tile.click()
                
                # Warte, bis die Artikelseite vollständig geladen ist (hier warten wir maximal 10 Sekunden)
                driver.implicitly_wait(10)
                
                # Extrahiere den Titel des Artikels
                title_element = driver.find_element("xpath", '/html/body/main/div/div/div[1]/header/div/div/div[2]/h1')
                article_title = title_element.text
                
                # Extrahiere das Datum des Artikels
                date_element = driver.find_element("xpath",'/html/body/main/div/div/div[1]/div[2]/div[1]/p[1]')
                article_date = date_element.text
                
                # Extrahiere den Inhalt des Artikels
                content_element = driver.find_element("xpath",'/html/body/main/div/div/div[2]/div/div[1]/div/p')
                article_content = content_element.text
                
                # Gib die extrahierten Informationen aus
                print("Titel:", article_title)
                print("Datum:", article_date)
                print("Inhalt:", article_content)
                print("\n")
            except StaleElementReferenceException:
                # If the element is stale, re-find the tiles
                tiles = driver.find_elements("xpath", tiles_selector)
                continue
                
            finally:
                # Gehe zurück zur Hauptseite, um das nächste Tile zu laden
                driver.back()
                
                # Warte, um sicherzustellen, dass die Hauptseite vollständig geladen ist
                time.sleep(1)
    except StaleElementReferenceException:
        # If the element is stale, re-find the tiles
        continue
        

# Schließe den Browser
driver.quit()
```
