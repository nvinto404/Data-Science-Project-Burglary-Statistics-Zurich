```python
import requests
from bs4 import BeautifulSoup

main_url = 'https://www.zh.ch/de/news-uebersicht.html?organisation=organisationen%253Akanton-zuerich%252Fsicherheitsdirektion%252Fkantonspolizei&fullText=Einbruch&page=1&orderBy=relevance'

main_response = requests.get(main_url)
main_soup = BeautifulSoup(main_response.text, 'html.parser')

tiles_selector = 'ul.mdl-news-list__items li a'
tiles = main_soup.select(tiles_selector)

# Durch alle Tiles iterieren
for tile in tiles:
    tile_url = tile['href']
    
    tile_response = requests.get(tile_url)
    tile_soup = BeautifulSoup(tile_response.text, 'html.parser')
    
    # Titel des Artikels extrahieren
    title_element = tile_soup.select_one('h1.atm-heading')
    article_title = title_element.get_text(strip=True) if title_element else "Titel nicht gefunden"
    
    # Datum des Artikels extrahieren
    date_element = tile_soup.select_one('div.mdl-lead_section__date time')
    article_date = date_element.get_text(strip=True) if date_element else "Datum nicht gefunden"
    
    # Inhalt des Artikels extrahieren
    content_element = tile_soup.select_one('div.mdl-news-article__content p')
    article_content = content_element.get_text(strip=True) if content_element else "Inhalt nicht gefunden"
    
    print("Titel:", article_title)
    print("Datum:", article_date)
    print("Inhalt:", article_content)
```

```python
import pandas as pd
import spacy
from collections import Counter

# Laden des deutschen Sprachmodells
nlp = spacy.load("de_core_news_sm")

directory = 'Cleaned_CSV_files'
current_directory = os.getcwd()
csv_path = os.path.join(current_directory, directory, "merged_ZurichNews.csv")

df = pd.read_csv(csv_path)

def extract_keywords(article):
    doc = nlp(article)
    tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct]
    # Extrahiere nur Wörter mit mindestens 3 Zeichen.
    keywords = [token for token in tokens if len(token) > 3]
    return keywords

all_keywords = []
for article in df["Tile-Text"]:
    keywords = extract_keywords(article)
    all_keywords.extend(keywords)

keyword_counter = Counter(all_keywords)

print("Die 10 am häufigsten vorkommenden Schlüsselwörter:")
for word, freq in keyword_counter.most_common(10):
    print(word, "-", freq)
```

```python
import pandas as pd
from textblob import TextBlob

directory = 'Cleaned_CSV_files'
current_directory = os.getcwd()
csv_path = os.path.join(current_directory, directory, "merged_ZurichNews.csv")
df = pd.read_csv(csv_path)

def analyze_sentiment(article):
    blob = TextBlob(article)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        return "positiv"
    elif sentiment_score < 0:
        return "negativ"
    else:
        return "neutral"

df["Sentiment"] = df["Tile-Text"].apply(analyze_sentiment)
print(df[["Tile-Text", "Sentiment"]])

# Zähle die Anzahl der positiven, negativen und neutralen Sentiments
positiv_count = (df["Sentiment"] == "positiv").sum()
negativ_count = (df["Sentiment"] == "negativ").sum()
neutral_count = (df["Sentiment"] == "neutral").sum()

print("Anzahl der positiven Sentiments:", positiv_count)
print("Anzahl der negativen Sentiments:", negativ_count)
print("Anzahl der neutralen Sentiments:", neutral_count)


```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Laden der Daten
directory = 'Cleaned_CSV_files'
current_directory = os.getcwd()
csv_path = os.path.join(current_directory, directory, "merged_ZurichNews.csv")
df = pd.read_csv(csv_path)

# Initialisierung des Sentiment-Analysators
analyzer = SentimentIntensityAnalyzer()

# Funktion zur Analyse des Sentiments für einen Artikel
def analyze_sentiment_vader(article):
    sentiment = analyzer.polarity_scores(article)
    if sentiment['compound'] >= 0.05:
        return "positiv"
    elif sentiment['compound'] <= -0.05:
        return "negativ"
    else:
        return "neutral"

# Anwenden der Sentiment-Analyse auf jeden Artikel
df["Sentiment_VADER"] = df["Tile-Text"].apply(analyze_sentiment_vader)

# Ausgabe der Ergebnisse
print(df[["Tile-Text", "Sentiment_VADER"]])

# Zähle die Anzahl der positiven, negativen und neutralen Sentiments
positiv_count = (df["Sentiment_VADER"] == "positiv").sum()
negativ_count = (df["Sentiment_VADER"] == "negativ").sum()
neutral_count = (df["Sentiment_VADER"] == "neutral").sum()

print("Anzahl der positiven Sentiments:", positiv_count)
print("Anzahl der negativen Sentiments:", negativ_count)
print("Anzahl der neutralen Sentiments:", neutral_count)
```

Der Unterschied in den Ergebnissen zwischen TextBlob und VADER kann auf die unterschiedlichen Ansätze und Funktionsweisen der beiden Sentimentanalyse-Tools zurückzuführen sein.

TextBlob verwendet einen einfacheren Ansatz, der auf einem lexikonbasierten Verfahren beruht. Es weist jedem Wort in einem Text eine Polarität zu und berechnet dann den Durchschnitt der Polaritäten aller Wörter im Text. Dadurch kann TextBlob die Stimmung eines Textes grob als positiv, negativ oder neutral klassifizieren.

VADER hingegen verwendet ein lexikonbasiertes und regelbasiertes Verfahren, das speziell für die Analyse von sozialen Medien entwickelt wurde. Es berücksichtigt verschiedene linguistische Eigenschaften wie Grossbuchstaben, Emojis und Satzstrukturen, um ein feineres Verständnis des Sentiments zu erlangen. Der "compound"-Score in VADER gibt eine Gesamtbewertung des Sentiments eines Textes basierend auf verschiedenen Aspekten wie Polarität, Intensität und Emotionen.

Themenmodellierung mit LDA

```python
from gensim import corpora, models
import pandas as pd
import spacy

# Laden des deutschen Sprachmodells
nlp = spacy.load("de_core_news_sm")

directory = 'Cleaned_CSV_files'
current_directory = os.getcwd()
csv_path = os.path.join(current_directory, directory, "merged_ZurichNews.csv")
df = pd.read_csv(csv_path)

# Die Funktion entfernt auch Stoppwörter und Satzzeichen und konvertiert die Tokens in Kleinbuchstaben. 
# Durch token.is_stop und token.is_punct usw. da diese keine semantische bedeutung haben.
def tokenize(article):
    doc = nlp(article)
    tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct]
    return tokens


# Die Tokenisierungsfunktion wird auf die Spalte "Tile-Text" angewendet, die den Text der Artikel enthält. 
df["Tokens"] = df["Tile-Text"].apply(tokenize)

# Erstellen des Wörterbuchs und des Korpus für LDA. Das Wörterbuch (Dictionary) wird erstellt, um jedem eindeutigen Token im Text einen numerischen Wert zuzuweisen. 
# Diese numerischen Werte dienen als IDs für die Tokens. Der Bag-of-Words-Korpus wird erstellt, indem für jeden Artikel im DataFrame eine Liste von Tupeln erstellt wird. 
#Jedes Tupel enthält die ID eines Tokens und die Häufigkeit, mit der dieses Token im entsprechenden Artikel vorkommt. 
dictionary = corpora.Dictionary(df["Tokens"])
corpus = [dictionary.doc2bow(tokens) for tokens in df["Tokens"]]

lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)

print("Die Themen in den Artikeln sind:")
for idx, topic in lda_model.print_topics(-1):
    print("Thema {}: {}".format(idx, topic))
```

Die Gewichtungen in einem Thema wird durch probabilistische Modelle wie Latent Dirichlet Allocation (LDA) berechnet. Je höher das Gewicht eines Wortes in einem Thema ist, desto stärker ist die Assoziation dieses Wortes mit diesem Thema. Die Zahlen vor den Wörtern geben die Gewichtung dieser Wörter innerhalb des Themas an.  Die Themenmodellierung bietet Methoden zum automatischen Organisieren, Verstehen, Suchen und Zusammenfassen grosser elektronischer Archive.
Sie kann bei Folgendem helfen:
* Entdeckung der versteckten Themen in der Sammlung.
* Klassifizierung der Dokumente in die entdeckten Themen.
* Verwendung der Klassifizierung zur Organisation/Zusammenfassung/Suche der Dokumente. LDA ist eine der beliebtesten Methoden zur Themenmodellierung. Jedes Dokument besteht aus verschiedenen Wörtern, und jedes Thema hat auch verschiedene Wörter, die zu ihm gehören. Das Ziel von LDA ist es, Themen zu finden, zu denen ein Dokument gehört, basierend auf den darin enthaltenen Wörtern. (Kulshrestha, 2021)

Quelle: Kulshrestha, R. (2021, 10. Dezember). A Beginner’s Guide to Latent Dirichlet Allocation(LDA). Medium. https://towardsdatascience.com/latent-dirichlet-allocation-lda-9d1cd064ffa2
